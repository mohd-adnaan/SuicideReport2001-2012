```python
df.rename(columns={'Type': 'Reasons'}, inplace=True)
df.head()
```

```python
def remove_uppercase(row):
  state = row['State'].strip()
  state = state.upper()
  return state
state_series = df.apply(remove_uppercase, axis=1)
df['State']=state_series

#Now we'll replace the otehr type of repeated data as discused above
df['State'].replace("A&N ISLANDS" , "A & N ISLANDS", inplace = True )
df['State'].replace("D&N HAVELI","D & N HAVELI", inplace = True)
df['State'].replace("DELHI (UT)", "DELHI", inplace =True)

#unique data length
"The Total number of State + UT : " + str(len(df['State'].unique()))
```

```python
suicide_reasons = list(df["Reasons"].unique())
suicide_reasons
```

```python
victim_counts = []
for reason in suicide_reasons:
  victims = len(df[df["Reasons"] == reason])
  victim_counts.append(victims)
```

```python
victim_df = pd.DataFrame({"Reason":suicide_reasons,"Victims Count":victim_counts})
victim_df
```

```python
# Sort the victim_df by "Victims Count" in descending order
sorted_major_reasons = victim_df.sort_values(by='Victims Count', ascending=False)

# Select the top N reasons you want to display (for example, top 10)
top_reasons = sorted_major_reasons.head(10)

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c2f0c2','#ff6666','#c2f0f0','#ffdb58']
import matplotlib.pyplot as plt
plt.pie(top_reasons["Victims Count"], radius=1.8, labels=top_reasons["Reason"], shadow=True, autopct='%1.1f%%', colors=colors)
plt.title("Distribution of Major Suicide Reasons", bbox={'facecolor':'#FFD1D1'})  # Use a shade of red
plt.axis('equal')


plt.show()
```

```python
# Assuming victim_df is already defined and sorted by "Victims Count"
top_10_reasons = victim_df.head(10)

# Define a custom color palette for the bars
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'lightcoral', 'lightblue', 'lightpink', 'gold', 'lightseagreen', 'lightgrey']

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_reasons["Reason"], top_10_reasons["Victims Count"], color=colors)
plt.xlabel('Reasons')
plt.ylabel('Victims Count')
plt.xticks(rotation=90)
plt.title('Top 10 Reasons for Suicide')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("Suicide Report 2001-12.csv")

# Group by Year and Gender and sum the total count
gender_counts = df.groupby(['Year', 'Gender'])['Total'].sum().reset_index()

# Pivot the data for plotting
pivot_df = gender_counts.pivot(index='Year', columns='Gender', values='Total')

# Plotting the figure
fig, ax = plt.subplots(figsize=(10, 6))

pivot_df.plot(kind='bar', stacked=True, ax=ax, colormap='Set1')
plt.xlabel('Year')
plt.ylabel('Total Suicides')
plt.title('Total Suicides by Gender and Year')
plt.legend(title='Gender')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("Suicide Report 2001-12.csv")

# Group by Year and Gender and sum the total count
gender_counts = df.groupby(['Year', 'Gender'])['Total'].sum().reset_index()

# Pivot the data for plotting
pivot_df = gender_counts.pivot(index='Year', columns='Gender', values='Total')

# Plotting the figure
fig, ax = plt.subplots(figsize=(12, 6))

pivot_df.plot(kind='line', ax=ax, marker='*', colormap='Set1')
plt.xlabel('Year')
plt.ylabel('Suicides')
plt.title('Total Suicides by Gender and Year')
plt.legend(title='Gender')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("Suicide Report 2001-12.csv")

# Group by State, Year, and sum the total count
state_year_counts = df.groupby(['State', 'Year'])['Total'].sum().reset_index()

# Get unique states
states = state_year_counts['State'].unique()

# Loop through each state and create a plot
for state in states:
    state_data = state_year_counts[state_year_counts['State'] == state]
    plt.figure(figsize=(10, 6))
    plt.plot(state_data['Year'], state_data['Total'], marker='o', label=f'Suicides in {state}')
    plt.xlabel('Year')
    plt.ylabel('Total Suicides')
    plt.title(f'Total Suicides in {state}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()  # Display the plot
```
